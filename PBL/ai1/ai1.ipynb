```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import MinMaxScaler
```

```python
# 데이터 로드 
df = pd.read_csv('diabetes.csv')

# 원본 데이터 상위 5행 확인 
df.head()
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 0값을 NaN으로 변환 (0은 결측치로 간주)
zero_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[zero_cols] = df[zero_cols].replace(0, np.nan)

# 변환 후 결측치 개수 확인
df.isnull().sum()
```

```python
# 결측치를 각 컬럼의 평균값으로 대체
for col in zero_cols:
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)

df.isnull().sum()
```

```python
# 이상치 처리 
## SkinThickness, Insulin 컬럼에서 상위 1% 값을 이상치로 간주
outlier_cols = ['SkinThickness', 'Insulin']

for col in outlier_cols:
    q99 = df[col].quantile(0.99)
    normal_mean = df.loc[df[col] <= q99, col].mean() # 이상치 제외한 평균값 계산
    df.loc[df[col] > q99, col] = normal_mean # 이상치를 평균값으로 대체

    print(f"{col} 이상치 처리 완료")
    print(f"{col} 99% 분위수 기준값: {q99:.2f}")
    print(f"{col} 대체에 사용한 평균값: {normal_mean:.2f}")
```

```python
# Age 컬럼 정규화
## MinMaxScaler를 사용하여 Age 컬럼을 0~1 범위로 변환
scaler = MinMaxScaler()
df[['Age']] = scaler.fit_transform(df[['Age']])

df[['Age']].head()
```

```python
# EDA 1 : 각 열의 결측치 개수 출력
df.isnull().sum()
```

```python
# EDA 2 : Outcome 별 Gloucose 평균 출력
## Outcome이 0인 그룹과 1인 그룹의 Glucose 평균을 비교
glucose_mean_by_outcome = df.groupby('Outcome')['Glucose'].mean()

print(f"Outcome 0 그룹의 Glucose 평균: {glucose_mean_by_outcome[0]:.2f}")
print(f"Outcome 1 그룹의 Glucose 평균: {glucose_mean_by_outcome[1]:.2f}")
```
